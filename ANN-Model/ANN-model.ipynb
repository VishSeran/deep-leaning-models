Let's start by randomly initializing the weights and the biases in the network. We have 6 weights and 3 biases, one for each node in the hidden layer as well as for each node in the output layer.


```python
import numpy as np

weights = np.around(np.random.uniform(size=6), decimals=2)#Initilize weights
bias = np.around(np.random.uniform(size = 3), decimals=2)#Initialize bias

print("weights: \n", weights)
print("bias: \n", bias)
```

Now that we have the weights and the biases defined for the network, let's compute the output for a given input, $x_1$ and $x_2$.


```python
x_1 = 0.5
x_2 = 0.85

print("x1 is {} and x2 is {}".format(x_1, x_2))
```

Let's start by computing the weighted sum of the inputs, $z_{1, 1}$, at the first node of the hidden layer.

```python
z_11 = x_1 * weights[0] + x_2 * weights[1]  + bias[0]
print('The weighted sum of the inputs at the first node in the hidden layer is {}'
      .format(np.around(z_11, decimals=4)))
```

Next, let's compute the weighted sum of the inputs, $z_{1, 2}$, at the second node of the hidden layer. Assign the value to **z_12**.


```python
z_12 = x_1 * weights[2] + x_2 *  weights[3] + bias[1]
print('The weighted sum of the inputs at the second node in the hidden layer is {}'.format(np.around(z_12,decimals=4)))
```

Next, assuming a sigmoid activation function, let's compute the activation of the first node, $a_{1, 1}$, in the hidden layer.


```python
a_11 = 1.0/(1.0 + np.exp(-z_11))

print('The activation of the first node in the hidden layer is {}'
      .format(np.around(a_11,decimals=4)))
```

```python
a_12 = 1.0/(1.0 + np.exp(-z_12))
print('The activation of the second node in the hidden layer is {}'
      .format(np.around(a_12,decimals=4)))
```

Now these activations will serve as the inputs to the output layer. So, let's compute the weighted sum of these inputs to the node in the output layer. Assign the value to **z_2**.


```python
z_2 = a_11 * weights[4] + a_12 * weights[5] + bias[2]
print('The weighted sum of the inputs at the first node in the second hidden layer is {}'.format(np.around(z_2,decimals=4)))
```

```python
a_2 = 1.0/(1.0 + np.exp(-z_2))
print('The activation of the second node in the hidden layer is {}'
      .format(np.around(a_2,decimals=4)))
```

## Build a Neural Network


Let's start by formally defining the structure of the network.


```python
# number of inputs
n =2

#number of hodden layers
hidden_layers = 2

#number of nodes each layer
m = [2,2]

#number of outputs layer nodes
output_nodes = 1
```

Now that we defined the structure of the network, let's go ahead and inititailize the weights and the biases in the network to random numbers. In order to be able to initialize the weights and the biases to random numbers, we will need to import the **Numpy** library.


```python
import numpy as np

num_nodes_previous = n # number of nodes in the previous layer
network = {} # network define as empty dict

for layer in range (hidden_layers+1):
    if (layer == (hidden_layers)):
        layer_name = 'output'
        num_nodes = output_nodes
    
    else:
        layer_name = 'layer_{}'.format(layer+1)
        num_nodes = m[layer]
        
    # initialize weights and biases associated with each node in the current layer
    network[layer_name] = {}
    
    for node in range(num_nodes):
        node_name = 'node_{}'.format(node+1)
        network[layer_name][node_name] = {
            'weight': np.around(np.random.uniform(size = num_nodes_previous),decimals=4),
            'bias': np.around(np.random.uniform(size = 1),decimals=4)
        }
    num_nodes_previous = num_nodes

print(network)
```

So for any layer with n nodes, there are always exactly n biases — one per node, no sharing.
The bias is like each node's personal adjustment knob — it shifts the output up or down independently of the inputs.

Awesome! So now with the above code, we are able to initialize the weights and the biases pertaining to any network of any number of hidden layers and number of nodes in each layer. But let's put this code in a function so that we are able to repetitively execute all this code whenever we want to construct a neural network.


```python
def initialize_network(num_inputs, num_hidden_layers, num_hidden_nodes, num_outputs):
    
    num_prev_nodes = num_inputs
    
    for layer in range(num_hidden_layers+1):
        if (layer == num_hidden_layers):
            layer_name = 'output'
            num_nodes = num_outputs
            
        else:
            layer_name = 'layer_{}'.format(layer+1)
            num_nodes = num_hidden_nodes[layer]
        
        network[layer_name] = {}
          
        for node in range(num_nodes):
            node_name = 'node_{}'.format(node + 1)
            network[layer_name][node_name] = {
                'weight': np.around(np.random.uniform(size=num_prev_nodes),decimals=4),
                'bias': np.around(np.random.uniform(size=1),decimals=4)
            }
        
        num_prev_nodes = num_nodes

    return network
            
```

```python
##initialize a samll network

small_network = initialize_network(5,3,[3,2,3],1)
small_network
```

## Compute Weighted Sum at Each Node
The weighted sum at each node is computed as the dot product of the inputs and the weights plus the bias. So let's create a function called *compute_weighted_sum* that does just that.


```python
def weighted_sum (inputs, weights, bias):
    return np.sum(inputs * weights) + bias
```

```python
# lets generate 5 inputs randomly
np.random.seed(12)
inputs = np.around(np.random.uniform(size=5), decimals=4)
print("Inputs of the network are {}".format(inputs))
```

#### Use the *compute_weighted_sum* function to compute the weighted sum at the first node in the first hidden layer.


```python
node_weights = small_network['layer_1']['node_1']['weight']
node_bias = small_network['layer_1']['node_1']['bias']

weight_sum = weighted_sum(inputs, node_weights,node_bias) 
print("The weighted sum at the first node in the hidden layer is {}".format(np.around(weight_sum[0],decimals=4)))
weight_sum
```

## Compute Node Activation


Recall that the output of each node is simply a non-linear tranformation of the weighted sum. We use activation functions for this mapping. Let's use the sigmoid function as the activation function here. So let's define a function that takes a weighted sum as input and returns the non-linear transformation of the input using the sigmoid function.


```python
def node_activation (node_weight_sum):
    return (1.0/(1.0 + np.exp(-1 * node_weight_sum)))
```

#### Use the *node_activation* function to compute the output of the first node in the first hidden layer.


```python
node_outputs = node_activation(weighted_sum(inputs, node_weights, node_bias))
print("The activation value at the first node in the hidden layer is {}".format(np.around(node_outputs[0],decimals=4)))
node_outputs
```
